```python
# exercise -1
import tensorflow as tf
#create a scalar
scalar = tf.constant(8)
#create a vector
vector = tf.constant([3,4])
#create a matrix
matrix = tf.constant([[2,1],
                      [6,8]])
#create a tensor
tensor = tf.zeros(5)
scalar, vector, matrix, tensor
```

```python
# Exercise - 2
#shape
shape = matrix.shape
#rank
rank = matrix.ndim
#size
size = tf.size(matrix)

shape, rank, size
```

```python
#Exercise - 3
#creating two random tensors betweeen 0 and 1 with shape [5,300]

tensor_1 = tf.random.uniform(shape=(5,300), minval = 0, maxval = 1)
tensor_2 = tf.random.uniform(shape=(5,300), minval = 0, maxval = 1)
tensor_1, tensor_2
```

```python
# Exercise - 4
#multiplication of the above two tensors

multiply_result = tf.matmul(tensor_1, tf.transpose(tensor_2))
multiply_result
```

```python
# Exercise - 5
# multiplication using the dot product

dot_product = tf.tensordot(tensor_1, tf.transpose(tensor_2), axes = 0)
dot_product
```

```python
# Exercise - 6
# Create a tensor with random values between 0 and 1 with shape [224, 224, 3].

tensor_3 = tf.random.uniform(shape=(224,224,3), minval = 0, maxval = 1)
tensor_3
```

```python
#Exercise - 7
#Find the min and max values of the tensor you created in 6 along the first axis.

maxi = tf.math.reduce_max(tensor_3, axis = 1)
mini = tf.math.reduce_min(tensor_3, axis = 1)

maxi, mini
```

```python
# Exercise - 9
#Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value.

dummy_tensor = tf.constant([4,1,14,72,91,24,43,98,21,34])
maximum_value = tf.argmax(dummy_tensor).numpy()
minimum_value = tf.argmin(dummy_tensor).numpy()
maximum_value, minimum_value
```
